```python
X = df[['male_height', 'male_weight']].iloc[: , :].values
```

```python
X_kmeans = X.copy()
inertia = []
for n in range(1 , 12):
    algorithm = KMeans(n_clusters=n)
    algorithm.fit(X_kmeans)
    inertia.append(algorithm.inertia_)
```

```python
plt.figure(figsize = (15 ,6))
plt.plot(np.arange(1 , 12) , inertia , 'o')
plt.plot(np.arange(1 , 12) , inertia , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.show()
```

```python
model = KMeans(n_clusters=4).fit(X_kmeans)
labels = model.labels_
centroids = model.cluster_centers_
```

```python
h = 0.02
x_min, x_max = X_kmeans[:, 0].min() - 1, X_kmeans[:, 0].max() + 1
y_min, y_max = X_kmeans[:, 1].min() - 1, X_kmeans[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
plt.figure(figsize=(15 , 7))
plt.clf()
Z = Z.reshape(xx.shape)
plt.imshow(Z , interpolation='nearest', 
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap = plt.cm.Pastel2, aspect = 'auto', origin='lower')

plt.scatter(x='male_height', y='male_weight', data=df , c=labels , s=100 )
plt.scatter(x = centroids[: , 0], y=centroids[: , 1], s=300 , c='red' , alpha=0.5)
plt.ylabel('male_weight'), plt.xlabel('male_height')
plt.show()
```
